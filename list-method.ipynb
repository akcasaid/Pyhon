```python
my_list = [1, 2, 3]
```

# Append


```python
# append(): Bir listeye öğe eklemek için kullanılır.
my_list.append(4) 
print (my_list)
```

# Extend

```python
#extend(): Bir listenin sonuna başka bir liste, tuple, set veya herhangi bir iterable'ın (tekrarlanabilir nesnenin) öğelerini ekler.
my_list = [1, 2, 3]
my_list.extend([4, 5])  # Listeyi [4, 5] ile genişletir.
print (my_list)
```

```python
'''
* append: Bu metot, tek bir öğeyi listeye sonundan ekler. Eğer eklenen öğe bir liste ise, bu liste tek bir öğe olarak eklenir ve böylece iç içe bir listeye sahip olursunuz.

* extend: Bu metot, bir listenin sonuna başka bir iterable'ın (liste, tuple, string vb.) tüm öğelerini ekler. Yani extend kullanıldığında, eklenen öğeler tek tek listeye dahil edilir ve sonuçta tek bir düz liste oluşur.
'''
```

# Insert

```python
my_list = [1, 2, 3]

print (my_list)
```

```python
# insert(): Belirli bir indekse öğe eklemek için kullanılır

my_list = [1, 2, 3]
my_list.insert(2, 3)
print (my_list)
```

# Remove

```python
#remove(): Listeden belirli bir öğeyi kaldırır. Öğe listede birden fazla kez bulunuyorsa, yalnızca ilk örneğini kaldırır.
my_list = [1, 2, 3, 2]
my_list.remove(2)  # İlk bulduğu '2' öğesini kaldırır.
print (my_list)
```

# Pop

```python
#pop(): Belirli bir indeksteki öğeyi kaldırır ve kaldırılan öğeyi döndürür. İndeks belirtilmezse, son öğeyi kaldırır.
my_list = [1, 2, 3]
removed_item = my_list.pop()  # Son öğeyi kaldırır ve döndürür.
print (my_list)
```

# Clear

```python
#clear(): Listenin tüm öğelerini kaldırır, böylece liste boşalır.
my_list = [1, 2, 3]
my_list.clear() 
print (my_list)
```

# Index

```python
#index(): Belirli bir öğenin ilk indeksini döndürür.
my_list = ['a', 'b', 'c', 'd']
index = my_list.index('b')  # 'b' öğesinin indeksini döndürür.
print (index)
```

# Count 

```python
# count(): Listedeki belirli bir öğenin sayısını döndürür
my_list = [1, 2, 2, 3, 2]
count = my_list.count(2)  # '2' öğesinin sayısını döndürür.
print (count)
```

# Short

```python
# sort(): Listeyi artan sırada sıralar. reverse=True argümanı verilerek azalan sırada sıralama yapılabilir.
my_list = [3, 1, 4, 1, 5]
my_list.sort()  # Listeyi sıralar.
print (my_list)
```

# Reserve

```python
#reverse(): Listenin öğelerinin sırasını tersine çevirir.
my_list = [1, 2, 3]
my_list.reverse()  # Listeyi ters çevirir.
print (my_list)
```

# Copy

```python
#copy(): Listenin bir kopyasını oluşturur.
my_list = [1, 2, 3]
new_list = my_list.copy()  # Listeyi kopyalar.
print (new_list)
```

# Len

```python
# len(): Bu, aslında bir liste metodundan ziyade Python'da yerleşik bir fonksiyondur ve bir listenin uzunluğunu, yani içerdiği öğe sayısını döndürür.
my_list = [1, 2, 3]
length = len(my_list)  # Listenin uzunluğunu döndürür.
print (length)
```

# Max - Min

```python
#max(): Yine bir liste metodundan ziyade Python'da yerleşik bir fonksiyondur ve bir listenin içindeki en büyük öğeyi döndürür.
#min(): Bu da Python'da yerleşik bir fonksiyondur ve bir listenin içindeki en küçük öğeyi döndürür.
my_list = [1, 2, 3]
max_value = max(my_list)  # Listenin en büyük öğesini döndürür.
print ( "Max değer: ", max_value)
min_value = min(my_list) 
print ( "Min değer: ", min_value)
```

# Sum

```python
my_list = [1, 2, 3]
total = sum(my_list)  # Listenin öğelerinin toplamını döndürür.
print ("Listenin toplam değer:" ,total)
```

# All

```python
#all(): Listenin tüm öğeleri True olarak değerlendiriliyorsa True döndürür. Boş bir liste için de True döndürür.
my_list = [True, 1, 'a']
all_true = all(my_list)  # Tüm öğeler True olarak değerlendiriliyorsa True döndürür.
print (all_true)
```

# Any

```python
#any(): Listenin herhangi bir öğesi True olarak değerlendiriliyorsa True döndürür. Boş bir liste için False döndürür.
my_list = [False, 0, '']
any_true = any(my_list)  # Herhangi bir öğe True olarak değerlendiriliyorsa True döndürür.
print (any_true)
```
